```python
#functions with parameters

def math_ops(a,b):
  print(a+b)
  print(a-b)
  print(a*b)
  print(a/b)

math_ops(1,2)
print("*" * 50)
math_ops(11,3)
```

```python
#positional based functions

def login(username,password):
  if username =="John" and password =="3212":
    print("Login success")
  else:
    print("Login Failed")

login("mike",6576)
login("john","5676")
login("John","3212")
login("3212","John")
```

```python
#without default arguments

def emp_info(name,email,location):
  print(f'Hi {name}, your email is {email} and location is {location}')

emp_info("Peter","peter@gmail.com","Ashford")
emp_info("sajee@gmail.com","sajee","Leeds")
```

```python
#with default arguments

def emp_info(name="Elliot",email="elliot@gmail.com",location="London"):
  print(f'Hi {name}, your email is {email} and location is {location}')

emp_info()
emp_info("user","user@gmail.com")
emp_info("arun@gmail.com","arun","Hyd")
```

```python
#with keyword argument -->

def emp_info(name,email,location="india"):
  print(f'Hi {name}, your email is {email} and location is {location}')

emp_info(name="nishanth",location="Gurgaon",email="nishanth@gmail.com")
emp_info(name="kayathiri",email="kayathiri@gmail.com")
```

```python
#Arbitary positional argurmnrt based(*args)

def add_unknown(*numbers):
  sum = 0
  for i in numbers:
    sum = sum + i
    print(sum)

add_unknown(1,3,5,7,9,11)
add_unknown(1,2,39)
```

```python
#Arbitary keyword argument based(^kwargs)

def profile(**info):
  print(info)

profile(name="ravi")
profile(id=101,name="mike",location="usa")
```

```python
#sample example for keyargs

def cred_transactions(**transactions):
  print(transactions)

  total = 0
  for i in transactions:
    total = total + transactions[i]
  print(f'You have done {len(transactions)} and total value of transactions is {total}')

cred_transactions(jan=100,feb=900,march=890,june=1234)
```

```python
def math_new(a,b,opr):
  if opr == "+":
    return a+b
  elif opr == "-":
    return a-b
  elif opr == "/":
    return a/b
  else:
    return "Invalid operation"

print(math_new(20,65,"-"))
print(math_new(50,12,"+"))
print(math_new(286,2,"/"))
```
